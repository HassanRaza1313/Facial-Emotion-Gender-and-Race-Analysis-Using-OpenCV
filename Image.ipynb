```python
import cv2
from deepface import DeepFace
import matplotlib.pyplot as plt
```

# Taking images as input

```python
img1 = cv2.imread("dp.jpg")
```

# Showing input image as BGR (Blue Green Red)

```python
plt.imshow(img1)
```

# Showing input image as RGB (Red Green Blue)

```python
plt.imshow(cv2.cvtColor(img1,cv2.COLOR_BGR2RGB))
```

# Analyzing image for facial expression, age, gender and race

```python
predictions1 = DeepFace.analyze(img1)
```

# Drawing ovelay around the face

```python
faceCascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

## converting BGR image to gray

```python
gray1 = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
faces1 = faceCascade.detectMultiScale(gray1,1.3,5)
```

## Drawing the rectangle

```python
for(x,y,w,h) in faces1:
    cv2.rectangle(img1,(x,y),(x+w,y+h),(0,255,0),2)
```

## Showing the overlay

```python
plt.imshow(cv2.cvtColor(img1,cv2.COLOR_BGR2RGB))
```

```python
font1 = cv2.FONT_HERSHEY_SIMPLEX
result1 = predictions1["dominant_emotion"] + "," + str(predictions1["age"]) + "," + predictions1["gender"] + "," + predictions1["dominant_race"] 

cv2.putText(img1,
            result1,
            (0,10),
            font1,0.5,
            (0,0,255),
            1,
            cv2.LINE_8
            )
```
